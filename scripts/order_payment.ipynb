```python
import pandas as pd
```

```python
orders = pd.read_csv("../raw/orders.csv")
orders.head()
```

```python
payments = pd.read_csv("../raw/payments.csv")
payments.head()
```

```python
order_payment = pd.merge(orders, payments, on = "order_id", how = 'left')
order_payment.head()
```

```python
order_payment = order_payment.drop(columns = ['payment_id'])
order_payment.head()
```

```python
order_payment.to_csv('../denormalised tables/order_payment.csv')
```
